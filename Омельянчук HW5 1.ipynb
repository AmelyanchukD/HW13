```python
df = pd.read_excel()
```

```python
# переименование столбцов (в нашем датасете в этом нет необходимости)
df = df.rename(columns=lambda c: c.lower().replace('-', '_').replace('/', '_').replace(' ', '_'))
```

```python
2. Cоздать столбцы с месяцем и днем (наименование дня недели)
```

```python
# Cоздаeм столбец с месяцем
df['month'] = df.transaction_date.dt.month 
```

```python
# Cоздаeм столбец с днем (наименованием дня недели)
df['weekday'] = df.transaction_date.dt.day_name()
df.head()
```

```python
3.  Посчитать количество транзакций по месяцам
```

```python
# выводим количество транзакций по месяцам
df.groupby('month')['transaction_qty'].sum()
```

```python
 Посчитать количество транзакций по дням недели в разрезе product_category 
```

```python
# выводим количество транзакций по дням недели в разрезе product_category 
pd.crosstab(df.weekday,
            df.product_category,
            values=df.transaction_qty,
            aggfunc='sum',
            margins=True,
            margins_name="Total")
```

```python
Посчитать количество транзакций в зависимости store_location и product_category (pivot table)
```

```python
# выводим количество транзакций по дням недели в зависимости store_location и product_category, используя (pivot table)
pd.pivot_table(df,
               index=["store_location", "product_category"],
               values=["transaction_qty"],
               aggfunc=[np.sum])
```

```python
Посчитать total_revenue (то есть добавить такой столбец)
```

```python
# Cоздаeм столбец и посчитаем total_revenue
df['total_revenue'] = df.transaction_qty*df.unit_price
df.head()
```

```python
 Посчитать суммарный total_revenue в разрезе store_location и в разрезе product_category (pivot_table)
```

```python
# выводим суммарный total_revenue в разрезе store_location и в разрезе product_category,используя (pivot table)
pd.pivot_table(df,
               index=["store_location", "product_category"],
               values=["total_revenue"],
               aggfunc=[np.sum])
```

```python
Вывести самый дешевый и дорогой product_type по unit_price в разрезе product_category
```

```python
# Выводим самый дорогой product_type по unit_price в разрезе product_category
pricemax = pd.pivot_table(df,
               index=["product_category", "product_type"],
               values=["unit_price"],
               aggfunc= [np.max])
pricemax.columns = ['most_expensive']
pricemax = pricemax.sort_values(by=['product_category', 'most_expensive'], ascending=[True, False])
pricemax = pricemax.groupby('product_category').nth(0)
pricemax
```

```python
# Выводим самый дешевый product_type по unit_price в разрезе product_category
pricemin = pd.pivot_table(df,
               index=["product_category", "product_type"],
               values=["unit_price"],
               aggfunc= [np.min])
pricemin.columns = ['the_cheapest']
pricemin = pricemin.sort_values(by=['product_category', 'the_cheapest'], ascending=[True, True])
pricemin = pricemin.groupby('product_category').nth(0)
pricemin
```

```python
# выводим самый дешевый и дорогой product_type по unit_price в разрезе product_category в один датасет
price = pd.concat([pricemin, pricemax], axis=1)
price
```

```python
Посчитать total revenue по месяцам 
```

```python
# Считаем total revenue по месяцам
df.groupby('month')['total_revenue'].sum()
```

```python
Найти второй по популярности продукт по transaction_qty в разрезе product_category
(то есть мы ищем второй по популярности product type в разрезе каждого product_category по количеству transaction_qty)
```

```python
# Группировка данных и подсчет общего количества продаж каждого типа продукта внутри каждой категории
popular_product_types = df.groupby(['product_category', 'product_type'])['transaction_qty'].sum().reset_index()

# Сортировка результатов по убыванию количества и выбор второго по популярности типа продукта
popular_product_types = popular_product_types.sort_values(by=['product_category', 'transaction_qty'], ascending=[True, False])
popular_product_types = popular_product_types.groupby('product_category').nth(1)
popular_product_types.columns = ['product_category', 'second_most_popular_product_type', 'transaction_qty']
popular_product_types



```
