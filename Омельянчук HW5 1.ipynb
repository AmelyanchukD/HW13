```python
df = pd.read_excel('Coffee Shop Sales.xlsx')
```

```python
# переименование столбцов (в нашем датасете в этом нет необходимости)
df = df.rename(columns=lambda c: c.lower().replace('-', '_').replace('/', '_').replace(' ', '_'))
```

2. Cоздать столбцы с месяцем и днем (наименование дня недели)

```python
# Cоздаeм столбец с месяцем
df['month'] = df.transaction_date.dt.month 
```

```python
# Cоздаeм столбец с днем (наименованием дня недели)
df['weekday'] = df.transaction_date.dt.day_name()
df.head()
```

3.  Посчитать количество транзакций по месяцам

```python
# выводим количество транзакций по месяцам
df.groupby('month')['transaction_qty'].sum()
```

 Посчитать количество транзакций по дням недели в разрезе product_category 

```python
# выводим количество транзакций по дням недели в разрезе product_category 
pd.crosstab(df.weekday,
            df.product_category,
            values=df.transaction_qty,
            aggfunc='sum',
            margins=True,
            margins_name="Total")
```

```python
# Cоздаeм столбец и посчитаем total_revenue
df['total_revenue'] = df.transaction_qty*df.unit_price
df.head()
```
